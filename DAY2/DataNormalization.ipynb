Data Normalization:

    Normalization is a specific form of feature scaling that transforms the range of features to a standard scale. 

Normallized Data:

    Normalized data enhances model performance and improves the accuracy of a model.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

```

```python
# Load your dataset
df = pd.read_csv(r'C:\Users\ayush\OneDrive\Desktop\coding\100dy for ml\Day1\train.csv')
```

```python
df.head()
```

USING MIN_MAX SCALAR:

    Min-Max Scaling transforms features by scaling them to a fixed range, usually [0, 1].

```python
scaler = MinMaxScaler()

# Assuming you want to normalize all numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Apply Min-Max Scaling
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

Standardization (Z-score Normalization):

    Standardization scales features to have a mean of 0 and a standard deviation of 1.

```python
scaler = StandardScaler()

# Assuming you want to standardize all numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Apply Standardization
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
df.head()
```

Min-Max Scaling: Use when you want to transform data into a specific range. It’s useful for algorithms that rely on distances, like k-NN or gradient descent-based algorithms.

Standardization: Use when your data has a Gaussian distribution and you want to handle outliers better. It’s generally preferred for most machine learning algorithms.
