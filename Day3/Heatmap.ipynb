HeatMap:

    A heatmap is a powerful visualization tool in machine learning that uses color to represent data values in a matrix format. Itâ€™s particularly useful for identifying patterns, correlations, and anomalies in datasets.
    Heatmaps organize data in a grid, with different colors or shades indicating different levels of the data's magnitude.
    

Heatmaps are best suited for datasets where you can represent values as colors, typically continuous or discrete numerical data. 

However, you can also use them for categorical data that has been quantified or summarized

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd .read_csv(r"C:\Users\ayush\OneDrive\Desktop\coding\100dy for ml\Day1\train.csv")
```

```python
df.head()
```

```python
# Select only numerical columns for correlation
numerical_df = df.select_dtypes(include=['float64', 'int64'])
```

```python
print(numerical_df)
```

```python
# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Heatmap of Correlation Matrix for Numerical Features')
plt.show()
```
