ScatterPlot:

    A scatter plot, also known as a scatter graph or scatter diagram, is a type of graph used to display the relationship between two quantitative variables. Each point on the scatter plot represents an observation in the dataset, with the position of the point determined by the values of the two variables.
    Scatter plots are particularly useful for identifying patterns, trends, and potential outliers in data


When To Use ScatterPlot:

    To detect patterns, trends, or clusters in your data. For instance, you might notice that data points form a certain shape or trend, indicating a specific relationship.
    When comparing two variables to see how one variable changes with respect to another.

Real Time Example

        Regression Analysis
        Outlier Dection

Scatter plots are best suited for numerical relationships. Avoid using them for categorical data unless you encode the categories into numbers

#Plotting Scatter Plot On AGE VS FARE

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.6, c='blue', edgecolors='black')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare')
plt.grid(True) 
plt.show()
```

ADDING REGRESSION LINE TO SHOW THE RELATIONSHIP BETWEEN AGE AND FARE

```python
import numpy as np
```

```python
# Clean the data: Drop NaNs and remove infinite values
df_filtered = df[['Age', 'Fare']].dropna()
```

```python
# Ensure no infinite values in 'Age' or 'Fare'
df_filtered = df_filtered[(df_filtered['Age'] != np.inf) & (df_filtered['Fare'] != np.inf)]
```

```python
x = df_filtered['Age']
y = df_filtered['Fare']

# Calculate the regression line
m, b = np.polyfit(x, y, 1)  # m = slope, b = intercept

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.6, c='blue', edgecolors='black', label='Data Points')
# Plot the regression line
plt.plot(x, m * x + b, color='red', linestyle='-', linewidth=2, label='Regression Line')

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare with Regression Line')
plt.grid(True) 
plt.legend()
plt.show()
```
