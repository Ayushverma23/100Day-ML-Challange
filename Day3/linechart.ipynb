Line Chart

    A line chart or line graph, also known as curve chart, is a type of chart that displays information as a series of data points called 'markers' connected by straight line segments.
    Line charts are particularly useful in machine learning for visualizing trends, patterns, and relationships between variables over time or sequential data. 

When To Use Line Chart:

    Time Series Analyisis
    Tracking Model performance
    Visualization Relationship

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ayush\OneDrive\Desktop\coding\100dy for ml\Day1\train.csv")
```

```python
# Count unique values for each column
unique_counts = df.nunique()
```

```python
columns = unique_counts.index.tolist()

# Plot the line chart
plt.figure(figsize=(12, 8))
plt.plot(columns, unique_counts.values, marker='o', linestyle='-', color='skyblue')
plt.xlabel('Columns')
plt.ylabel('Count of Unique Values')
plt.title('Count of Unique Values for Each Column')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels
plt.grid()  # Optional: Add grid for better readability
plt.show()
```
