A histogram is a graphical representation of the distribution of numerical data. It uses bars to show the frequency of data points within certain ranges, known as bins.

Construction:

    Bins: Divide the entire range of values into a series of intervals.
    Frequency: Count how many values fall into each interval.
    Bars: Draw rectangles (bars) for each bin, with the height representing the frequency of data points in that bin.

Used for continious and numerical value

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\ayush\OneDrive\Desktop\coding\100dy for ml\Day1\train.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Passenger Ages')
plt.grid(axis='y', alpha=0.75)  # Optional: Add grid for better readability
plt.show()
```

```python
# Plot the histogram for the 'Fare' column
plt.figure(figsize=(10, 6))
plt.hist(df['Fare'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Passenger Fares')
plt.grid(axis='y', alpha=0.75)  # Optional: Add grid for better readability
plt.show()
```
