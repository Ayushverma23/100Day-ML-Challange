Bar Chart:

    A bar chart (also called bar graph) is a chart that represents data using bars of different heights.
    The bars can be two types – vertical or horizontal. It doesn’t matter which type you use.

When To Use Bar Chart:

    Comparing discrete categories (e.g., sales by region).
    Showing changes over time if categories are discrete (e.g., monthly revenue).
    When you need to show how large data changes over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\ayush\OneDrive\Desktop\coding\100dy for ml\Day1\train.csv')
```

Printing all the columns in dataset using bar chart

```python
# Count unique values for each column
unique_counts = df.nunique()
```

```python
# Plot a bar chart
plt.figure(figsize=(12, 8))
plt.bar(unique_counts.index, unique_counts.values, color='red')
plt.xlabel('Columns')
plt.ylabel('Count of Unique Values')
plt.title('Count of Unique Values for Each Column')
plt.xticks(rotation=360, ha='right')  # Rotate labels for better readability
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

Every bar in different colors

```python
import numpy as np
```

```python
cmap = plt.get_cmap('tab20', len(unique_counts))
colors = [cmap(i) for i in range(len(unique_counts))]

# Convert Index to a list for plotting
columns = unique_counts.index.tolist()

# Plot the bar chart
plt.figure(figsize=(12, 8))
bars = plt.bar(columns, unique_counts.values, color=colors)
plt.xlabel('Columns')
plt.ylabel('Count of Unique Values')
plt.title('Count of Unique Values for Each Column')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to fit labels
plt.tight_layout()

# Optional: Add legend
plt.legend(handles=[bars[i] for i in range(len(bars))],
           labels=columns,
           bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

Bar Grpah For Specific columns

```python
#for survie column
survival_counts = df['Survived'].value_counts()

# Plot a bar chart
plt.figure(figsize=(8, 6))
plt.bar(survival_counts.index, survival_counts.values, color=['red', 'green'])
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Survival Status')
plt.xticks(ticks=[0, 1], labels=['Did Not Survive', 'Survived'])
plt.show()
```

```python
#For Sex column
gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Gender')
plt.show()
```
